```python
import spacy
```

```python
nlp = spacy.load('en_core_web_lg', disable=['parser', 'tagger', 'ner'])
```

```python
nlp.max_length = 1198623
```

```python
def separate_punct(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
d = read_file('../UPDATED_NLP_COURSE/06-Deep-Learning/moby_dick_four_chapters.txt')
tokens = separate_punct(d)
tokens
```

```python
len(tokens)
```

Here we provide a sequence of 25 words and have our network predict the 26th word.

Here 25 words is long enough to grab the structure of a sentence but not short enough where you are missing general context

```python
train_len = 25+1
text_sequences = []

for i in range(train_len, len(tokens)):
    seq = tokens[i-train_len: i]
    
    text_sequences.append(seq)
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
import numpy as np
```

```python
sequences = np.array(sequences)
sequences
```

```python
from keras.utils import to_categorical
```

```python
X= sequences[:,:-1]
X
```

```python
y = sequences[:, -1]
y
```

```python
y = to_categorical(y, num_classes=vocabulary_size +1)
```

```python
seq_len= X.shape[1]
seq_len
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
```

```python
def create_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, seq_len, input_length=seq_len))
    
    model.add(LSTM(50, return_sequences=True))
    model.add(LSTM(50))
    
    model.add(Dense(50,activation='relu'))
    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    model.summary()
    
    return model
```

```python
model = create_model(vocabulary_size+1, seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(X, y, batch_size=128, epochs=300, verbose=1)
```

```python
model.save('my_mobydick_model.h5')
```

```python
from keras.utils.data_utils import pad_sequences
```

```python
# model => trained model
# tokenizer => tokenizer has the knowledge about the vocabulary like what id number goes in for what word
# seq_len => lenght of the sequence

# seed_text => some seed word that we are going to start of with. This is robust enough to have shorter seed 
# text or longer seed text than the sequence length, But for best purposes or best we should try to make these
# essentially equal to each other. You should make the seed text the same length as what was trained on,
# otherwise you will have to pad it

# num_gen_words => number of words we want to generate. If we have more number of words to be 
# generated i.e. bigger than sequence length, then we end up predicting the word using predicted words 
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    
    output_text = []
    # initial seed sequence
    input_text = seed_text
    
    for i in range(num_gen_words):
        #Here i am going to first take the input text string and encode it to be a sequence. This is what we 
        # did earlier as well where we encoded raw text into sequence of numbers
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        # If my seed text is too short or too long I may have to pad it, I may need to cut it off or I may 
        # need to add it. Here we truncate the very first word after the model prediction to match seq_length.
        # If we have 
        pad_encoded= pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        # Here we are predicting the class probabilities for each word, it will go through the entire vocabulary
        # and assign a probability to each word and the most likely next word gets the highest probability
        predict_x = model.predict(pad_encoded, verbose=0)
        pred_word_ind = np.argmax(predict_x,axis=1)[0]
        
        # Deprecated
        # pred_word_ind = model.
        # pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]

        # Here we return the word in the index from the tokenizer vocabulary 
        pred_word = tokenizer.index_word[pred_word_ind]
        
        input_text += ' '+pred_word
        output_text.append(pred_word)
        
    return ' '.join(output_text)
```
